```python
import json

# Load the JSON file - use the CORRECT filename
with open('sample.json', 'r') as file:
    data = json.load(file)

print("✅ File loaded successfully!")
print(f"Type of data: {type(data)}")
print(f"Number of records: {len(data)}")

# Let's see the first record to understand the structure
print("\nFirst record:")
print(json.dumps(data[0], indent=2) if isinstance(data, list) else json.dumps(data, indent=2))
```

```python
# Extract all order IDs and their total values
order_ids_and_totals = []

for customer in data:
    for order in customer['orders']:
        order_info = {
            'order_id': order['order_id'],
            'total': order['total']
        }
        order_ids_and_totals.append(order_info)

print("Order IDs and their total values:")
for item in order_ids_and_totals:
    print(f"Order ID: {item['order_id']}, Total: ${item['total']}")
```

```python
import pandas as pd
df_flat = pd.DataFrame()
for customer in data:
    for order in customer['orders']:
        for item in order['items']:
            temp_df = pd.DataFrame([{
                'customer_id': customer['customer_id'],
                'order_id': order['order_id'],
                'product': item['product'],
                'quantity': item['quantity'],
                'total': order['total']
            }])
            df_flat = pd.concat([df_flat, temp_df], ignore_index=True)
print(df_flat)
```

```python
df_flat.to_json('processed_orders.json', orient='records', indent=2)
from google.colab import files
files.download('processed_orders.json')
```
